```python
from matplotlib import pyplot as plt
tabMeteo3 = [((2022,1,20), 2, 16, 0), ((2022,1,21), 1, 13, 0.2), ((2022,1,23), 6, 19, 0.6), ((2022,1,24), 3, 18, 0.8),((2022,2,20), 6, 19, 0.2), ((2022,2,24), 3, 18, 0.2), ((2022,2,28), 3, 18, 0.2)]
def medias(tabMeteo):
    res = []
    for dia in tabMeteo3:
        tmin=dia[1]
        tmax=dia[2]
        data=dia[0]
        media=(tmin+tmax)/2
        tuplo=(data,media)
        res.append(tuplo)
    return res
def guardaTabMeteo(t, fnome):
    f=open(fnome,"w")
    for data,tmin,tmax,prec in t:
        ano, mes, dia=data
        f.write(f"{ano}-{mes}-{dia}; {tmin};{tmax};{prec}\n")
    f.close()
    return
guardaTabMeteo(tabMeteo3, "meteorologia.txt")
def carregaTabMeteo(fnome):
    res = []
    file=open(fnome,"r")
    for line in file:
        line=line.strip()
        campos=line.split(";")
        data,tmin, tmax,prec=campos
        ano,mes,dia=data.split("-")
        tuplo=((int(ano),int(mes),int(dia)),float(tmin),float(tmax),float(prec))
        res.append(tuplo)
        print(line)
    file.close()
    return res
tabMeteo2 = carregaTabMeteo("meteorologia.txt")
def minMin(tabMeteo):
    minima=10000
    for dado in tabMeteo3:
        tempmin=dado[1]
        if tempmin<minima:
            minima=tempmin
    return minima
def amplTerm(tabMeteo):
    res = []
    for dado in tabMeteo3:
        tmax=dado[2]
        tmin=dado[1]
        data=dado[0]
        diferença=tmax-tmin
        tuplo=(data,diferença)
        res.append(tuplo)
    return res
def maxChuva(tabMeteo):
    maximo=0
    for dado in tabMeteo3:
        prec=dado[3]
        data=dado[0]
        if prec>maximo:
            maximo=prec
            tuplo=(data,maximo)
    return tuplo
def diasChuvosos(tabMeteo, p):
    resultado=[]
    for dado in tabMeteo3:
        prec=dado[3]
        data=dado[0]
        if prec>p:
            tuplo=(data,prec)
            resultado.append(tuplo)
    return resultado
def maxPeriodoCalor(tabMeteo, p):
        diasconsecutivos=0
        
        for dado in tabMeteo3:
            prec=dado[3]
            if prec<p:
                diasconsecutivos+=1    
            else:
                diasconsecutivos=0
        return diasconsecutivos
def grafTabMeteo(t):
    x=[f"{data[0]}-{data[1]}-{data[2]}" for data,tmin,tmax,prec in t]
    ytmin=[tmin for data,tmin,tmax,prec in t ]
    ytmax=[tmax for data,tmin,tmax,prec in t]
    yprec=[prec for data,tmin,tmax,prec in t]
    plt.plot(x,ytmin, label= "Temperatura minima", color="indigo",marker="o")
    plt.plot(x,ytmax, label= "Temperatura Máxima",color="fuchsia",marker="o")
    plt.legend()
    plt.grid()
    plt.title("Tabela Metereológica")
    plt.xticks(rotation=45)
    plt.show()
    plt.bar(x, yprec, color="pink")
    plt.show()
    return

escolha=-1
while escolha!=0:
        escolha=int(input(f"""\n=== Aplicação Meteorológica, a tabela meteriológica é: {tabMeteo3} ===
            --> 1 - Temperatura média de cada dia
            --> 2 - Guardar uma tabela meteorológica num ficheiro de texto
            --> 3 - Carregar uma tabela meteorológica de um ficheiro de texto
            --> 4 - Temperatura mínima mais baixa registada na tabela
            --> 5 - Amplitude térmica
            --> 6 - Dia em que a precipitação registada teve o seu valor máximo
            --> 7 - Função que recebe uma tabela meteorológica e um limite "p" e retorna uma lista correspondente aos dias em que a precipitação foi superior a "p"
            --> 8 - Função que recebe uma tabela meteorológica e um limite "p" e retorna o maior número consecutivo de dias com precipitação abaixo desse limite
            --> 9 - Mostrar gráfico"""))
            
            
        if escolha==1:
            print(medias(tabMeteo3))
        elif escolha==2:
            print("Tabela guardada")
        elif escolha==3:
            print(tabMeteo3)
        elif escolha==4:
            print(f"A temperatura minima é: {minMin(tabMeteo3)}")
        elif escolha==5:
            print(amplTerm(tabMeteo3))
        elif escolha==6:
            print(maxChuva(tabMeteo3))
        elif escolha==7:
            p=float(input("Deseja observar a precipitação superior a que valor?"))
            print(diasChuvosos(tabMeteo3,p))
        elif escolha==8:
            p=float(input("Deseja observar a precipitação inferior a que valor?"))
            print(maxPeriodoCalor(tabMeteo3,p)) 
        elif escolha==9:
            grafTabMeteo(tabMeteo3)
        else:
            print("Valor não aceite")
    
        

```
