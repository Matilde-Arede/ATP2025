```python
import random


fósforos=21
soma=int(fósforos)
resposta=input("O jogo começa com 21 fósforos. Deseja começar (Sim ou não)?")
if resposta=="sim":
    while soma>0:
        jogada=input("Quantos fósforos deseja retirar?")
        número=int(jogada)
        if número==1 or número==2 or número==3 or número==4:
            soma-=número
            print(f"Retirou {número} fósforos. Restam {soma} fósforos!")
        else: 
            print("Erro! Escolha um número entre 1 e 4!")
            continue
        if soma==1:
            print("Ganhaste!")
            break
        num=5-número
        soma-=num
        if num>soma:
            num=min(4,soma)
        print(f"O computador retirou {num} fósforos. Restam {soma} fósforos!")
        if soma==1:
            print("Perdeste")
            break
elif resposta=="não":
    computador=random.randint(1,4)
    soma-=computador
    print(f"O computador retirou {computador} fósforos. Restam {soma} fósforos!")
    while soma>=1:
        jogo=input("Quantos fósforos deseja retirar?")
        numerojogo=int(jogo)
        if numerojogo==1 or numerojogo==2 or numerojogo==3 or numerojogo==4:
            soma-=numerojogo
            print(f"Retirou {numerojogo} fósforos. Restam {soma} fósforos!")
        else: 
            print("Erro! Escolha um número entre 1 e 4!")
            continue
        if soma==1:
            print("Ganhaste!")
            break 

        if soma%5!=1:
            computador=(soma-1)%5
            soma-=computador
            print(f"O computador retirou {computador} fósforos. Restam {soma} fósforos!")
        
        else:
            computador=random.randint(1,min(4,soma-1))
            soma-=computador
            print(f"O computador retirou {computador} fósforos. Restam {soma} fósforos!")
        if soma==1:
            print("Perdeste!")
            break
else:
    print("Resposta inválida! Recomece e responda apenas com sim ou não!")
    
            
    
   
```
