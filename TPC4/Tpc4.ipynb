```python
sala1=["Sala 1",40,[],"13 going on 30"]
sala2=["Sala 2",30,[],"Harry Potter"]
sala3=["Sala 3",60,[],"Barbie"]
sala4=["Sala 4",45,[],"A noiva cadáver"]

Cinema=[sala1,sala2,sala3,sala4]
import random



def listar(cinema):
    print("Em exibição")
    for sala in cinema:
        print(f"{sala[0]}:{sala[3]}")
    return


def disponivel(Cinema, filme, lugar):
    for sala in Cinema:
        if filme == sala[3]:
            if lugar in sala[2]:
                return False
            else: 
                return True
    return False
def vendeBilhete(Cinema,filme,modo):
    if modo=="Não": 
        for sala in Cinema:
            if filme==sala[3]:
                res = ""
                while res != "aceite":
                    lugar=random.randrange(1,sala[1]+1) 
                    if disponivel(Cinema, filme, lugar): 
                        sala[2].append(lugar)
                        print(f"O lugar {lugar} foi selecionado. Bom filme!!")
                        print(sala)
                        res = "aceite"
                    else:
                        res = "recusado"
        

    if modo=="Sim":
    
        for sala in Cinema:
            if filme==sala[3]:       
                lugar = int(input("Escolha o lugar que quer"))
                if disponivel(Cinema,filme,lugar):
                    sala[2].append(lugar)
                    print(f"O lugar {lugar} foi selecionado. Bom filme!")
                    print(sala) 
                else:
                    print("Este lugar já está ocupado, selecione outro!")

def listarDisponibilidades(Cinema):
    print("Estes são os filmes disponiveis no cinema:")
    for sala in Cinema:
        x= sala[1] - len(sala[2]) 
        print(f"{sala[0]}, com {x} lugares disponiveis.")
    return    
        



def inserirSala( cinema, sala, lotação, filme ):
    sala=[sala,lotação,[],filme]
    
    
    if sala not in cinema:
        cinema.append(sala)
    print("Sala escolhida!")
    print(f"{sala[0]}: lotação de {sala[1]} lugares. Filme que está em exibição:{sala[3]}")
    



menu=-1
print("""Olá! Selecione uma opção do menu: "
      (1) Listar todos os filmes em exibição;
      (2) Verificar a disponibilidade da sala e do lugar;
      (3) Comprar bilhete, somente após verificar a disponibilidade;
      (4) Listar os lugares restantes em cada sala;
      (5) Adiciona uma sala ao cinema;
      (6) Fechar o menu.""")
while menu !=6:
    menu=int(input("Escolhe a tua opção."))

    if menu==1:
        listar(Cinema)

    elif menu ==2:
        print(disponivel(Cinema,input("Escolhe o filme deseja ver"),int(input("Escolha um lugar da sala"))))

    elif menu ==3:
        vendeBilhete(Cinema,input("Escolha o filme que deseja ver"), input("Deseja escolher o lugar?(Sim ou Não)"))

    
    elif menu ==4:
        listarDisponibilidades(Cinema)

    elif menu ==5:
        inserirSala(Cinema,input("Qual o nome da nova sala?"),int(input("Qual a lotação da sala?")),input(("Qual o filme em exibição?")))
    
print("Obrigada!")
```
